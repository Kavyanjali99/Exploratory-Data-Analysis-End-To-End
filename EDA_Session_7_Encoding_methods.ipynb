**one hot encoder**

- One hot means at a time only one will **ON** and another is **OFF**

- **ON** represent with 1

- **OFF** represnt with 0

- one hot encoder creates new columns

- for example case statu has two unique lables

- Number of new columns = Number of unique lables

- case staus has two unique lables

    - Certified

    - Denied

- It will create two new columns

    - Case_status_Certified

    - Case_status_Denied

- Original column as prefix

|case_status|case_status_Certified|case_status_Denied
|-|-|-|
|Certified|1|0|
|Denied|0|1|
|Denied|0|1|
|Certified|1|0|

**Advantages**

- Generally ML models expect independance between input variables

- There is No relation between the variables

- Because of One hot at a time only one will be On , other will Off

- This indicates the New columns are independant each other

- This independance behaviour also called as **Orthogonality**

- Independant means No relation

- No relation means perpendicular each other

- Perpendicular means 90 degrees phase shift

- 90degrees means orthogonal

**Disadvanatage**

- it will create more columns

- more columns means more data

- More data more complexity w.r.t time and Infra

- This is called as **curse of Dimensionality**
